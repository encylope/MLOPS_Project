# Credit Card Fraud Detection — EDA
Exploratory Data Analysis for the Kaggle creditcardfraud dataset.

**Outputs used in MLOps pipeline:**
- Feature baseline statistics → `data/features/feature_baseline.json`
- Class distribution → logged to MLflow
- Drift detection reference values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/creditcard.csv')
print(df.shape)
df.head()
```

```python
# Class distribution
print('Fraud rate:', df['Class'].mean())
df['Class'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.show()
```

```python
# Baseline statistics for drift detection
import json
feature_cols = [c for c in df.columns if c != 'Class']
baseline = {}
for col in feature_cols:
    baseline[col] = {
        'mean': float(df[col].mean()),
        'std': float(df[col].std()),
        'min': float(df[col].min()),
        'max': float(df[col].max()),
        'p25': float(df[col].quantile(0.25)),
        'p75': float(df[col].quantile(0.75)),
    }
print(json.dumps(baseline['Amount'], indent=2))
```

```python
# Correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Amount distribution by class
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df[df['Class']==0]['Amount'].hist(bins=50, ax=axes[0])
axes[0].set_title('Legitimate — Amount')
df[df['Class']==1]['Amount'].hist(bins=50, ax=axes[1], color='red', alpha=0.7)
axes[1].set_title('Fraud — Amount')
plt.tight_layout()
plt.show()
```
